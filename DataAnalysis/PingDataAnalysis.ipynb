```python
import matplotlib.pyplot as plt
```

```python
gt = [5,10,20,30,40,50,60,80,100,120,140,160,180,200,220,250,270,300]
ping = [5,10,20,30,40,48,61,80,99,119,139,159,180,200,220,249,269,299]
```

```python
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 12}

plt.rc('font', **font)

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.plot(range(1,len(gt)+1),ping,'-o',label='PING')
plt.plot(range(1,len(gt)+1),gt,'-o',label='Ground Truth',alpha=0.6)
plt.xticks (range(1,len(gt)+1))
plt.yticks(gt)
plt.grid(linestyle='dotted')
plt.xlabel("Number of tirals")
plt.ylabel("Distance (cm)")

plt.legend()
```
